```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('Sample data.csv')
df
```

```python
# Created to track how many orders weree attempted before getting delivered.
df['order_Reattempt']= np.where(df['delivery_attempted'].notna(),1,0)
df
```

```python
# Created to check how many orders were delivered directly without getting cancelled after delivery.
df['is_Delivered']=np.where(df['LATESTORDERSTATUS']=='DELIVERED',1,0)
df
```

```python
df.columns
```

```python
# Created to track how many orders were cancelled
df['is_cancelled']= np.where(df['pendencyFlag']== 'cancelled',1,0)
df
```

```python
# Created to trackorders which are pending
df['pending']=np.where(df['pendencyFlag']=='pending',1,0)
df
```

```python
# Created to track the orders asking to be returned after delivery
df['Post_delivery_returns']= np.where(df['pendencyFlag']=='post_Delivery_Return',1,0)
df
```

```python
# Created to track the first TAT Adherence
df['Formatted_created_date'] = pd.to_datetime(df['created_ts_IST']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['Formatted_created_date'] = pd.to_datetime(df['Formatted_created_date'])
df['firstTAT']=pd.to_datetime(df['firstTAT']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['firstTAT']=pd.to_datetime(df['firstTAT'])
```

```python
# Created to track the difference between First TAT and created date
df['create_to_f_tat_diff'] = np.where(df['Formatted_created_date'].notna() & df['firstTAT'].notna(),
np.maximum ((df['firstTAT'] - df['Formatted_created_date']).dt.total_seconds() / 60,0),np.nan)
df
```

```python
# Created the bucket for first TAT and Created TAT diffarence
df['Created_to_F_TAT_Diff_0_to_2_hrs']=np.where(df['create_to_f_tat_diff']<121,1,0)
df
```

```python
#Created the buckeat for first TAT and created TAT differnce
df['Created_to_F_TAT_Diff_2_to_4_hrs']=np.where((df['create_to_f_tat_diff']<120) & (df['create_to_f_tat_diff']<241),1,0)
df
```

```python
# Created the bucket for first TAT and Created TAT differnece
df['four_8hrs']=np.where((df['create_to_f_tat_diff']>=241) & (df['create_to_f_tat_diff']<=480),1,0)
df
```

```python
# Created the bukets for firts TAT and Created TAT differnce
df['eight_24hrs']=np.where((df['create_to_f_tat_diff']>=481) & (df['create_to_f_tat_diff']<=1440),1,0)
df
```

```python
# Created the bucket for first TAT and Created TAT difference
df['twentyfour_48hrs']=np.where((df['create_to_f_tat_diff']>=1441) &(df['create_to_f_tat_diff']<=2880),1,0)
df
```

```python
df['greaterthan_48hrs']=np.where((df['create_to_f_tat_diff']>=2881) & (df['create_to_f_tat_diff']<=5760),1,0)
df
```

```python
# Created to track the Orders whose terninal status is delivered and cancelled after delivery
df['NetDelivered']=np.where(df['delivered'].notna,1,0)
df
```

```python
df['delivered']= pd.to_datetime(df['delivered']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['firstTAT']= pd.to_datetime (df['firstTAT']).dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Created to track first TAT Adherence
df['First_SLA_Adherence'] = np.where(df['NetDelivered'] == 1, np.where(df['delivered'] <= df['firstTAT'],1,0),np.nan)
df
```

```python
df['delivered']= pd.to_datetime(df['delivered']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['secondTAT']= pd.to_datetime (df['secondTAT']).dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
# Created to track second TAT Adherence
df['Second_SLA_Adherence'] = np.where(df['NetDelivered'] == 1, np.where(df['delivered'] <= df['secondTAT'],1,0),np.nan)
df
```

```python
df['delivered']= pd.to_datetime(df['delivered']).dt.strftime('%Y-%m-%d %H:%M:%S')
df['latestTAT']= pd.to_datetime (df['latestTAT']).dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
#Created to track Latest TAT Adherence
df['Latest_SLA_Adherence'] = np.where(df['NetDelivered'] == 1, np.where(df['delivered'] <= df['latestTAT'],1,0),np.nan)
df
```

```python
#Created to track total orders count
df['Count']=np.where(df['orderid'].notna,1,0)
df
```

```python
# Created to track where the actual pendency is
df['pre_store_pending']=np.where((df['pendencyFlag']=='pending') & (df['pendencyFlag']=='pre-verification'),1,0)
df
```

```python
# Created to track where the actual pendency is
df['prebill']=np.where((df['pendencyFlag']=='pending') & (df['pendencyFlag']=='pre-Billing'),1,0)
df
```

```python
# Created to track where the actual pendency is
df['postShip_Pending']= np.where((df['pendencyFlag']=='pending') &(df['pendencyType']=='in-Transit'),1,0)
df
```

```python
df['Store_Pushed']= np.where(df['SHOPID'].notna(),1,0)
df
```

```python
# Created to tarck where the actual pandency is
df['DSP_Assigned']=np.where(df['forwardDSP'].notna(),1,0)
df
```

```python
df['verified']=pd.to_datetime(df['verified'])
df['booked']=pd.to_datetime(df['booked'])
```

```python
# Time taken for verfication
df['verified_time_min']=np.where(df['verified'].notna()&df['booked'].notna(),
np.maximum((df['verified']-df['booked']).dt.total_seconds()/60,0),np.nan)
df
```

```python
df['last_shc']= pd.to_datetime(df['last_shc'])
df['billed']= pd.to_datetime (df['billed'])
```

```python
# Created to track how much time order took for billing once it landed on store
df['Billed_time_min'] = np.where(df['billed'].notna() & df['last_shc'].notna(),
np.maximum((df['billed'] - df['last_shc']).dt.total_seconds() / 60,0),np.nan)
df
```

```python
df['firstpacked']=pd.to_datetime(df['firstpacked'])
df['billed']=pd.to_datetime(df['billed'])
```

```python
# Created to track how much time order took for rider assigment
df['DSP_Allocation_In_Min'] = np.where(df['billed'].notna() & df['firstpacked'].notna(),
np.maximum((df['firstpacked'] - df['billed']).dt.total_seconds() / 60,0),np.nan)
df
```

```python
df['rider_reached_store'] = pd.to_datetime(df['rider_reached_store'])
df['last_ofd'] = pd.to_datetime(df['last_ofd'])
```

```python
df['RiderAssing_in_min']=np.where (df['rider_reached_store'].notna(),
np.maximum((df['last_ofd']- df['rider_reached_store']).dt.total_seconds()/60,0),np.nan)
df
```

```python
df['delivered']=pd.to_datetime(df['delivered'])
df['lastshipped']=pd.to_datetime(df['lastshipped'])
```

```python
#Created time in mintues using delivery and lastshipped timestamps
df['RiderDelivery_In_Min'] = np.where(df['delivered'].notna() & df['lastshipped'].notna(),
np.maximum ((df['delivered'] - df['lastshipped']).dt.total_seconds() / 60,0),np.nan)
df
```

```python
df['firstshipped']=pd.to_datetime(df['firstshipped'])
df['first_rts']=pd.to_datetime(df['first_rts'])
```

```python
#Created to track how much time order took to get picked up rider
df['Pickup_In_Min'] = np.where(df['firstshipped'].notna() & df['first_rts'].notna(),
np.maximum ((df['firstshipped'] - df['first_rts']).dt.total_seconds() / 60,0),np.nan)
df
```

```python
df['delivered']=pd.to_datetime(df['delivered'])
df['billed']=pd.to_datetime(df['billed'])
```

```python
# Created time in hours using delievery and billed timestamps
df['O2D_In_Hour'] = np.where(df['delivered'].notna() & df['billed'].notna(),
np.maximum ((df['delivered'] - df['billed']).dt.total_seconds() / 3600,0),np.nan)
df
```

```python
df['last_ofd']=pd.to_datetime(df['last_ofd'])
df['rider_reached_store']=pd.to_datetime(df['rider_reached_store'])
```

```python
# Created store wait time in hours using last_ofd and riderreached store
df['store_wait_time'] = np.where(df['last_ofd'].notna() & df['rider_reached_store'].notna(),
np.maximum ((df['last_ofd'] - df['rider_reached_store']).dt.total_seconds() / 3600,0),np.nan)
df
```
